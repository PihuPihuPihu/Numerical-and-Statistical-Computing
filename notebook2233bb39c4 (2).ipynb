```python
import numpy as np
import pandas as pd

print("FOR THE GIVEN QUESTION THE VALUES OF:")

# Step 1: Define x and f(x) values
x_arr = np.array([1, 2, 3, 4, 5])
print("x:")
print(x_arr)

fx_arr = np.array([1, 8, 27, 64, 125])  # f(x) = x^3
print("f(x):")
print(fx_arr)

# Step 2: Calculate h and u
dx = np.diff(x_arr)
print("Differences in x:", dx)

xn = 1.2  # Interpolation point near the beginning
h = dx[0]  # assuming uniform spacing
x0 = x_arr[0]
u = (xn - x0) / h
print("u =", u)

# Step 3: Build the forward difference table
n = len(fx_arr)
diff_table = np.zeros((n, n))
diff_table[:, 0] = fx_arr

for j in range(1, n):
    for i in range(n - j):
        diff_table[i][j] = diff_table[i + 1][j - 1] - diff_table[i][j - 1]

print("Forward Difference Table:")
print(pd.DataFrame(diff_table))

# Optional: Readable Table
column_names = ['f(x)'] + [f'Δ^{i}y' for i in range(1, n)]
diff_df = pd.DataFrame(diff_table[:, :n], columns=column_names)
diff_df.insert(0, 'x', x_arr)
print("\nRenamed Forward Difference Table:")
print(diff_df)

# Step 4: Newton Forward Interpolation Formula
def factorial(n):
    return 1 if n == 0 else n * factorial(n - 1)

f_xn = diff_table[0][0]
product_term = 1

for i in range(1, n):
    for j in range(i):
        product_term *= (u - j)
    f_xn += (product_term * diff_table[0][i]) / factorial(i)
    product_term = 1

print(f"\nInterpolated value at x = {xn} is: {f_xn}")
```
