```python
# BISECTION METHOD

print("GIVEN QUESTION: Find a root of the equation f(x) = x^3 - x - 2 using Bisection Method in the interval [1, 2]")

# Step 1: Define the function
def f(x):
    return x**3 - x - 2  # The equation given

# Step 2: Bisection method function
def bisection(a, b, tol=1e-6, max_iter=100):
    if f(a) * f(b) >= 0:
        print("Bisection method fails: f(a) and f(b) must have opposite signs.")
        return None

    print("\nIteration\t a\t\t b\t\t mid\t\t f(mid)")
    for i in range(max_iter):
        mid = (a + b) / 2
        f_mid = f(mid)

        print(f"{i+1}\t\t {a:.6f}\t {b:.6f}\t {mid:.6f}\t {f_mid:.6f}")

        if abs(f_mid) < tol or (b - a) / 2 < tol:
            print(f"\nRoot found at x = {mid:.6f} after {i+1} iterations")
            return mid

        if f(a) * f_mid < 0:
            b = mid
        else:
            a = mid

    print("\nMax iterations reached. Approximate root:", mid)
    return mid

# Step 3: Call the function with given interval
a = 1
b = 2
bisection(a, b)
```
