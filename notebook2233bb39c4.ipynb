```python
import numpy as np
import pandas as pd
import seaborn as sns

print("FOR THE GIVEN QUESTION THE VALUES OF:")

# Step 1: Define the x and f(x)
x_arr = np.array([1, 2, 3, 4, 5])
print("x:")
print(x_arr)

fx_arr = np.array([1, -1, 1, -1, 1])
print("f(x):")
print(fx_arr)

# Step 2: Compute step size and position
dx = np.diff(x_arr)
print("Differences in x:", dx)

xn = 3.5
h = dx[0]  # assuming uniform spacing
x0_index = len(x_arr) // 2  # middle index
x0 = x_arr[x0_index]
u = (xn - x0) / h
print("u =", u)

# Step 3: Create and populate the difference table
n = len(fx_arr)
diff_table = np.zeros((n, n))
diff_table[:, 0] = fx_arr

for j in range(1, n):
    for i in range(n - j):
        diff_table[i][j] = diff_table[i + 1][j - 1] - diff_table[i][j - 1]

print("Difference Table:")
print(pd.DataFrame(diff_table))

# Optional: Create a readable DataFrame for visualization
column_names = ['f(x)'] + [f'Δ^{i}y' for i in range(1, n)]
diff_df = pd.DataFrame(diff_table[:, :n], columns=column_names)
diff_df.insert(0, 'x', x_arr)
print("\nRenamed Difference Table:")
print(diff_df)

# Step 4: Apply Gauss Forward Interpolation Formula
def factorial(n):
    return 1 if n == 0 else n * factorial(n - 1)

# Use only the first row (starting from middle index) for Gauss Forward
f_xn = diff_table[x0_index][0]
product_term = 1

for i in range(1, n):
    if i % 2 == 1:
        idx = x0_index - (i // 2)
    else:
        idx = x0_index - (i // 2)

    for j in range(i):
        if j % 2 == 0:
            product_term *= (u - (j // 2))
        else:
            product_term *= (u + (j + 1) // 2)

    f_xn += (product_term * diff_table[x0_index - (i // 2)][i]) / factorial(i)
    product_term = 1

print(f"\nInterpolated value at x = {xn} is: {f_xn}")
```
