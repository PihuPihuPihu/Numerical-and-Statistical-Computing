```python
# NEWTON-RAPHSON METHOD

print("GIVEN QUESTION: Find a root of the equation f(x) = x^3 - x - 2 using Newton-Raphson Method with an initial guess x₀ = 1.5")

# Step 1: Define the function f(x) and its derivative f'(x)
def f(x):
    return x**3 - x - 2  # The equation given

def df(x):
    return 3*x**2 - 1  # Derivative of f(x)

# Step 2: Newton-Raphson method function
def newton_raphson(x0, tol=1e-6, max_iter=100):
    print("\nIteration\t xₙ\t\t f(xₙ)\t\t f'(xₙ)\t\t xₙ₊₁")
    
    for i in range(max_iter):
        fx = f(x0)
        dfx = df(x0)
        
        if dfx == 0:
            print("Derivative is zero. Newton-Raphson method fails.")
            return None
        
        # Newton-Raphson formula: xₙ₊₁ = xₙ - f(xₙ) / f'(xₙ)
        x1 = x0 - fx / dfx

        print(f"{i+1}\t\t {x0:.6f}\t {fx:.6f}\t {dfx:.6f}\t {x1:.6f}")
        
        if abs(x1 - x0) < tol:
            print(f"\nRoot found at x = {x1:.6f} after {i+1} iterations")
            return x1
        
        x0 = x1

    print("\nMax iterations reached. Approximate root:", x1)
    return x1

# Step 3: Call the function with initial guess x₀ = 1.5
x0 = 1.5
newton_raphson(x0)
```
