```python
# REGULA FALSI METHOD (FALSE POSITION)

print("GIVEN QUESTION: Find a root of the equation f(x) = x^3 - x - 2 using Regula Falsi Method in the interval [1, 2]")

# Step 1: Define the function
def f(x):
    return x**3 - x - 2  # The equation given

# Step 2: Regula Falsi method function
def regula_falsi(a, b, tol=1e-6, max_iter=100):
    if f(a) * f(b) >= 0:
        print("Regula Falsi method fails: f(a) and f(b) must have opposite signs.")
        return None

    print("\nIteration\t a\t\t b\t\t x\t\t f(x)")
    for i in range(max_iter):
        # Compute false position
        x = (a * f(b) - b * f(a)) / (f(b) - f(a))
        fx = f(x)

        print(f"{i+1}\t\t {a:.6f}\t {b:.6f}\t {x:.6f}\t {fx:.6f}")

        if abs(fx) < tol:
            print(f"\nRoot found at x = {x:.6f} after {i+1} iterations")
            return x

        if f(a) * fx < 0:
            b = x
        else:
            a = x

    print("\nMax iterations reached. Approximate root:", x)
    return x

# Step 3: Call the function with given interval
a = 1
b = 2
regula_falsi(a, b)
```
